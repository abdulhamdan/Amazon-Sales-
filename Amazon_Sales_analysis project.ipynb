```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Amazon Sales data.csv")
```

```python
df.head()
```

```python
df.drop(('Units Sold'),axis=1,inplace=True)
```

```python
df.drop(('Unit Cost'),axis=1,inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df
```

```python
df.value_counts()
```

```python
sns.distplot(df["Total Profit"])
```

```python
#Count of different Region
ax = sns.countplot(x = "Region",data=df)
sns.set(rc={'figure.figsize':(55,25)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# bar chart of total profit by country
ax = sns.barplot(x="Country", y="Total Profit", data=df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# pie chart for total profit by region

region_profit = df.groupby('Region')['Total Profit'].sum()

plt.figure(figsize=(15, 10))
plt.pie(region_profit, labels=region_profit.index, autopct='%1.1f%%')
plt.title('Total Profit by Region')
plt.show()
```

```python
df['Total Revenue'].value_counts()
```

```python
#Total Revenue by Region
total_revenue=df.groupby('Region')['Total Revenue'].sum()
plt.figure(figsize=(25,15))
plt.bar(total_revenue.index,total_revenue)
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.title('Total Revenue by Region')
plt.show()
```

```python
df.columns
```

```python
#count of item type
sns.countplot(x='Item Type',data=df)
```

```python
 #items sold in different countries with a graph

items_by_country = pd.crosstab(df['Country'], df['Item Type'])
items_by_country.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Items Sold')
plt.title('Items Sold in Different Countries')
plt.show()
```

```python
items_by_region = pd.crosstab(df['Region'], df['Item Type'])
items_by_region.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Items Sold')
plt.title('Items Sold in Different Region')
plt.show()
```
